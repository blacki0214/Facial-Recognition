```python
import cv2, numpy as np, tensorflow as tf

model = tf.keras.models.load_model("../models/liveness_detector_zalo.keras")
IMG_SIZE = (160,160)

def detect(frame):
    img = cv2.resize(frame, IMG_SIZE)
    img = np.expand_dims(img/255.0, axis=0)
    pred = model.predict(img)[0][0]
    return ("Real", pred) if pred >= 0.5 else ("Fake", pred)

cap = cv2.VideoCapture(0)
print("Press 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret: break
    label, score = detect(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    color = (0,255,0) if label == "Real" else (0,0,255)
    cv2.putText(frame, f"{label} ({score:.2f})", (20,40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
    cv2.imshow("Liveness Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): break

cap.release(); cv2.destroyAllWindows()
```
